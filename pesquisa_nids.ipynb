**# Pré-processamento e separação datasets**

*   NF-UNSW-NB15-v3
*   NF-ToN-IoT-v3
*   NF-BoT-IoT-v3
*   NF-CSE-CIC-IDS2018-v3

[Datasets](https://staff.itee.uq.edu.au/marius/NIDS_datasets/)



```python
# Funções auxiliares github
from nids_utils.preprocess import (
    aplicar_pca_datasets,
    converter_parquet,
    limpar_dataset,
    preparar_datasets_treino,
    separar_dataset,
)

DATASETS = [BOT, CICIDS, TON, UNSW]

train_size = 0.60
val_size = 0.15
n_components = 30
pca_batch_size = 1_000_000

for dataset in DATASETS:
    print(f"Processando {dataset['DATASET_KEY']}")
    converter_parquet(
        input_path=dataset["RAW_CSV"],
        output_path=dataset["RAW_PARQUET"],
    )

    # Seguido conforme artigo removendo valores ausentes
    limpar_dataset(
        input_path=dataset["RAW_PARQUET"],
        output_path=dataset["CLEAN_PARQUET"],
        numeric_features=NUMERIC_FEATURES,
        categorical_features=CATEGORICAL_FEATURES,
        target=TARGET,
    )

    # Separação treino, teste e validação
    train_file, val_file, test_file = separar_dataset(
        input_path=dataset["CLEAN_PARQUET"],
        output_path=dataset["SPLIT_PATH"],
        dataset_name=dataset["DATASET_KEY"],
        hash_features=HASH_FEATURES,
        target=TARGET,
        train_size=train_size,
        val_size=val_size,
        random_seed=SEED,
    )

    # Calcula média e desvio padrão só no treino para as colunas numéricas
    # Calcula quais categorias existem só no treino para as colunas categóricas
    # Aplica a mesma transformação nos três arquivos e aplica one-hot encoding
    train_processed, val_processed, test_processed = preparar_datasets_treino(
        train_path=train_file,
        val_path=val_file,
        test_path=test_file,
        output_path=dataset["SPLIT_PATH"],
        dataset_name=dataset["DATASET_KEY"],
        numeric_features=NUMERIC_FEATURES,
        categorical_features=CATEGORICAL_FEATURES,
        target=TARGET,
    )

    # Ajusta o PCA somente no treino e aplica para teste e validação
    aplicar_pca_datasets(
        train_path=train_processed,
        val_path=val_processed,
        test_path=test_processed,
        output_path=dataset["SPLIT_PATH"],
        dataset_name=dataset["DATASET_KEY"],
        artifact_path=dataset["ARTIFACT_PATH"],
        target=TARGET,
        n_components=n_components,
        batch_size=pca_batch_size,
    )
```

**# Treino**

```python
from nids_utils.train import ModelConfig, treinar_dataset
from tensorflow.keras.layers import Activation, BatchNormalization, Dense, Dropout, Input
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam

DATASETS = [BOT, CICIDS, TON, UNSW]

# Configurações de treinamento
config = ModelConfig(
    batch_size=2048,
    epochs=120,
    learning_rate=0.001,
    early_stop_patience=120,
    random_seed=SEED,
    clean_local_artifacts=True,
)

def nids_mlp(input_dim: int, config: ModelConfig) -> Sequential:
    model = Sequential(
        [
            Input(shape=(input_dim,)),
            Dense(64),
            BatchNormalization(),
            Activation("relu"),
            Dropout(0.1),
            Dense(32),
            BatchNormalization(),
            Activation("relu"),
            Dense(16),
            Dropout(0.1),
            BatchNormalization(),
            Activation("relu"),
            Dense(2, activation="softmax"),
        ]
    )

    model.compile(
        optimizer=Adam(
            learning_rate=config.learning_rate,
            beta_1=0.9,
            beta_2=0.999,
        ),
        loss="sparse_categorical_crossentropy",
        metrics=["accuracy"],
    )

    return model

for dataset in DATASETS:
    print(f"Treinando {dataset['DATASET_KEY']}")
    treinar_dataset(dataset=dataset, config=config, model_builder=nids_mlp)
```
